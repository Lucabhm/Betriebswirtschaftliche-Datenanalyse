# Analyse: Einfluss von Benzinpreisschwankungen auf die Inflationsrate (2019–2024)

## 1. Forschungsfrage

**Wie beeinflussen Schwankungen der Benzinpreise von 2019 bis 2024 die Entwicklung der Inflationsrate?**

Die Analyse erfolgt auf Basis von **Jahresdaten** für mehrere Länder. Ziel ist es, sowohl einen **länderübergreifenden Überblick** als auch eine **vertiefte Betrachtung einzelner Länder** zu ermöglichen.

---

## 2. Theoretischer Hintergrund

Benzinpreise beeinflussen die Inflation über mehrere Kanäle:

* Benzin ist Bestandteil des Verbraucherpreisindex (VPI)
* Steigende Benzinpreise erhöhen Transport- und Produktionskosten
* Unternehmen geben höhere Kosten an Konsument:innen weiter
* Besonders stark in Krisenzeiten (z. B. Pandemie, Energiekrise)

Daher wird ein **positiver Zusammenhang** zwischen Benzinpreisen und Inflationsrate erwartet.

### 3.1 Beschreibung der Datensätze

* **Inflationsrate**: jährliche Inflationsrate in Prozent
* **Zeitraum**: 2019–2024

## 6. Korrelationsanalyse

### 6.1 Ziel der Korrelationsanalyse

Untersuchung, ob ein statistischer Zusammenhang zwischen Benzinpreisen und Inflationsrate besteht.

### 6.2 Methodik

* Pearson-Korrelation (lineare Zusammenhänge)
* Optional: Spearman-Korrelation (robuster bei Ausreißern)
* Analyse **separat für jedes Land**

### 6.3 Interpretation

* Wertebereich: −1 bis +1
* Positiv: gleichgerichtete Entwicklung
* Negativ: gegenläufige Entwicklung

> Hinweis: Aufgrund der geringen Anzahl an Beobachtungen (n = 6) dienen die Ergebnisse primär der explorativen Analyse.
