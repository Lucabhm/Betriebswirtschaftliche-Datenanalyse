```python
import pandas as pd
import matplotlib.pyplot as plt

infl = pd.read_csv("../Datenbanken/global_inflation_data.csv")


country = "Germany"          
years = list(range(2010, 2025))
year_cols = [str(y) for y in years]

row = infl[infl["country_name"] == country].iloc[0]
values = row[year_cols].astype(float).values

plt.figure(figsize=(8, 4))
plt.plot(years, values, marker="o")
plt.xlabel("Jahr")
plt.ylabel("Inflationsrate (%)")
plt.title(f"Inflation in {country} ({years[0]}â€“{years[-1]})")
plt.grid(True, axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```
